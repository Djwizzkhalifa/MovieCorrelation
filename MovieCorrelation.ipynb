```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Now we need to read in the data
df = pd.read_csv('movies.csv', encoding = "ISO-8859-1")
```

```python
df.head()
```

```python

# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
df.dtypes
```

```python
df["budget"]=df["budget"].astype("int64")
#changing datatype of columns
df["gross"]=df["gross"].astype("int64")
```

```python
#creating a correct year column
df['yearcorrect']=df['released'].astype(str).str[:4]
```

```python
df.head()
```

```python
df= df.sort_values(by=['gross'], inplace=False,ascending=False)
df.head()
```

```python
pd.set_option('display.max_rows',None)
```

```python
#removing duplicates
#df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel("Gross Earnings")
plt.ylabel("Budget per Film")
plt.show()
```

```python
#plotting budget vs gross w/ seaborn
sns.regplot(x='budget',y='gross',data=df, scatter_kws={"color":"blue"},
line_kws={"color":"yellow"})
```

```python
#hence we see budget and gross are corelated
df.corr()
```

```python
correlation_matrix= df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation matrix for numeric features')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()

plt.show()
```

```python
df_num=df
for col_name in df_num.columns:
    if(df_num[col_name].dtype=='object'):
        df_num[col_name]= df_num[col_name].astype('category')
        df_num[col_name]= df_num[col_name].cat.codes
df_num.head()
```

```python
correlation_matrix= df_num.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)

plt.title('Correlation matrix for numeric features')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()

plt.show()
```

```python
correlation_mat = df_num.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs.head()
```

```python
sorted_pairs= corr_pairs.sort_values()
sorted_pairs.head()
```

```python
high_corr= sorted_pairs[(sorted_pairs)>0.5]
high_corr.head()
```
